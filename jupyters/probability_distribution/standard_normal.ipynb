# 標準正規分布

```python
from shiori.utils.viewer import set_plot_params, sns, plt
from shiori.utils.path import ProjectPath
from pathlib import Path
from scipy.stats import norm
import numpy as np
set_plot_params()
name = Path().cwd().name + "/standard_normal"
proj = ProjectPath(save_dst=name)
save_dir = proj.save_dir
save_dir.mkdir(parents=True, exist_ok=True)
plt.rcParams.update({
    "text.usetex": True,               # LaTeXを使う
    "font.family": "Arial",            # セリフ体
    "font.serif": ["Computer Modern"], # LaTeX標準フォント
})
```

```python
x = np.linspace(-4, 4, 500)

# 標準正規分布のPDF
pdf = norm.pdf(x, loc=0, scale=1)

# プロット
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label=r"$Z \sim \mathcal{N}(0,1)$", color="blue")
plt.fill_between(x, pdf, alpha=0.2, color="blue")
plt.title("Standard Normal Distribution PDF")
plt.xlabel(r"$z$")
plt.ylabel(r"$f(z)$")
plt.legend()
plt.grid(True)
plt.savefig(save_dir / "pdf.png", dpi=300, facecolor="white")
plt.show()
```

```python
# 標準正規分布のCDF
cdf = norm.cdf(x, loc=0, scale=1)

# プロット
plt.figure(figsize=(6,4))
plt.plot(x, cdf, label=r"$Z \sim \mathcal{N}(0,1)$", color="blue")
plt.title("Standard Normal Distribution CDF")
plt.xlabel(r"$z$")
plt.ylabel(r"$F(z)$")
plt.legend()
plt.grid(True)
plt.savefig(save_dir / "cdf.png", dpi=300, facecolor="white")
plt.show()
```

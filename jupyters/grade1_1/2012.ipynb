```python
def g1(x: float) -> float:
    return 3 * (1 - x) ** 2

def g2(x: float) -> float:
    return 6 * x * (1 - x)

def g3(x: float) -> float:
    return 3 * x ** 2
```

```python
x = np.linspace(0, 1, 100)
y1 = g1(x)
y2 = g2(x)
y3 = g3(x)
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label=r"$g_1(x) = 3(1-x)^2$")
plt.plot(x, y2, label=r"$g_2(x) = 6x(1-x)$")
plt.plot(x, y3, label=r"$g_3(x) = 3x^2$")
plt.legend()
plt.savefig(save_dir / "2012_1.png", dpi=300)
plt.show()
```
